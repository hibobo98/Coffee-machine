```python
class Coffee:

# 우선순위 : 돈  -> 재료 

    def __init__(self, bean =3 , water=3, sugar=3, frima=3, cup=3):
        self.bean = bean
        self.water = water
        self.sugar = sugar
        self.cup = cup
        self.frima = frima
        self.coffee = [self.bean, self.water, self.sugar,
                    self.frima, self.cup]
        self.price = [1000,1000,1500,2000] # 처음 가격 세팅
        self.wallet = 0 # 돈 저장 
        
        
    def makeandchange(self): # 돈, 커피 다 같이 처리
        if self.order == 1: # 블랙커피 0,1, 4
            self.wallet -= self.price[0]
            if self.wallet >= 0 :# 돈이 충분할 때
                if self.coffee[0] <= 0 or self.coffee[1] <= 0 or self.coffee[4] <= 0: # 재료 없으면
                    self.wallet+= self.price[0]
                    print("재료가 부족합니다. 안녕히 가세요.거스름돈 = {}".format(self.wallet))
                    self.wallet = 0
                    self.menu()
                else: # 재료 있으면
                    self.coffee[0] -= 1
                    self.coffee[1] -= 1
                    self.coffee[4] -= 1
                    print("남은 재료 확인용 : ", self.coffee) # 남은 재료 확인용
                    print(" 맛있게 드세요. 남은 돈 = {}".format(self.wallet))
                    self.wantmore() # 커피를 더 마실지.
                
            else: # 돈이 부족해!! 
                self.wallet+= self.price[0]
                print("돈이 부족합니다")
                self.menu()
                      
        elif self.order == 2: #설탕커피 0,1,2,4
            self.wallet -= self.price[1]
            if self.wallet >= 0 :# 돈 확인
                if self.coffee[0] <= 0 or self.coffee[1] <= 0 or self.coffee[2]<= 0 or self.coffee[4]<= 0: # 재료 없으면
                    self.wallet+= self.price[1]
                    print("재료가 부족합니다. 안녕히 가세요. 거스름돈 = {}".format(self.wallet))
                    self.wallet = 0
                    self.menu()
                      
                else: # 재료 있으면
                    self.coffee[0] -= 1
                    self.coffee[1] -= 1
                    self.coffee[2] -= 1
                    self.coffee[4] -= 1
                    print(self.coffee) # 남은 재료 확인
                    print(" 맛있게 드세요. 남은 돈 = {}".format(self.wallet))
                    self.wantmore() # 커피를 더 마실지.
        
            else: # 돈이 부족해!!  
                self.wallet+= self.price[1]
                print("돈이 부족합니다")
                self.menu()
                      
        elif self.order == 3: # 프림커피 0,1,3,4
            self.wallet -= self.price[2]
            if self.wallet >= 0 :# 돈 확인
                if self.coffee[0] <= 0 or self.coffee[1] <= 0 or self.coffee[3]<= 0 or self.coffee[4]<= 0: # 재료 없으면
                    self.wallet+= self.price[2]
                    print("재료가 부족합니다. 안녕히 가세요. 거스름돈 = {}".format(self.wallet))
                    self.wallet = 0
                    self.menu()
                else: # 재료 있으면
                    self.coffee[0] -= 1
                    self.coffee[1] -= 1
                    self.coffee[3] -= 1
                    self.coffee[4] -= 1
                    print(self.coffee) # 남은 재료 확인
                    print(" 맛있게 드세요. 남은 돈 = {}".format(self.wallet))
                    self.wantmore() # 커피를 더 마실지.
        
            else: # 돈이 부족해!!  
                self.wallet+= self.price[2]
                print("돈이 부족합니다")
                self.menu()

        elif self.order == 4: # 밀크커피 : 0,1,2,3,4
            self.wallet -= self.price[3]
            if self.wallet >= 0 :# 돈 확인
                if self.coffee[0] <= 0 or self.coffee[1] <= 0 or self.coffee[3]<= 0 or self.coffee[3]<= 0 or self.coffee[4]<= 0: # 재료 없으면
                    self.wallet+= self.price[3]
                    print("재료가 부족합니다. 안녕히 가세요. 거스름돈 = {}".format(self.wallet))
                    self.wallet = 0
                    self.menu()
                      
                else: #재료 있으면 
                    self.coffee[0] -= 1
                    self.coffee[1] -= 1
                    self.coffee[3] -= 1
                    self.coffee[4] -= 1
                    print(self.coffee) # 남은 재료 확인
                    print(" 맛있게 드세요. 남은 돈 = {}".format(self.wallet))
                    self.wantmore() # 커피를 더 마실지.
        
            else: # 돈이 부족해!!  
                self.wallet+= self.price[3]
                print("돈이 부족합니다")
                self.menu()
                


    def management(self):
        print("===== 자동판매기 관리 프로그램 =====")
        print("[1]재료보충 [2]금액조정 [3]작업종료 [4]영업종료")
        print("="*15)
        menu = int(input("메뉴 = "))
        if menu == 1:
            self.ingredient() # 재료보충
            
        elif menu == 2:
            self.update() # 금액조정
        
        elif menu ==3:
            print("작업을 종료합니다.")
            self.menu()
        elif menu ==4:
            print("영업종료 : 커피머신을 종료합니다. ")
            
        else:
            print("잘못된 번호입니다.")
            self.menu()
          
          
    def ingredient(self):
        print("===재료보충===")
        print("[1]전체 [2]커피 [3]물 [4]설탕 [5]프림 [6]컵 ")
        print("="*15)
        choice = int(input("선택= "))
        if choice == 1:
            self.coffee[0] = 3
            self.coffee[1] = 3
            self.coffee[2] = 3
            self.coffee[3] = 3
            self.coffee[4] = 3
            print(self.coffee) # 확인용
            print("모두 보충되었습니다.")
            self.menu() # 처음 메뉴로 돌아가
        elif choice == 2:
            self.coffee[0] = 3
            print(self.coffee) # 확인용
            print("커피가 보충되었습니다.")
            self.ingredient() # 재료보충 선택화면으로 돌아가 
            
        elif choice == 3:
            self.coffee[1] = 3
            print(self.coffee)
            print("물이 보충되었습니다.")
            self.ingredient()
          
        elif choice == 4:
            self.coffee[2] = 3
            print(self.coffee)
            print("설탕이 보충되었습니다.")
            self.ingredient()
          
        elif choice == 5 :
            self.coffee[3] = 3
            print(self.coffee)
            print("프림이 보충되었습니다.")
            self.ingredient()
          
        elif choice == 6:
            self.coffee[4] = 3
            print(self.coffee)
            print("컵이 보충되었습니다.")
            self.ingredient()
          
        else:
            print("잘못된 번호입니다.")
            self.ingredient() ##  다시 번호선택으로 돌아가기 
            
            
    def update(self): # 금액 조정 
        choice2 = int(input("수정할 음료 선택 [1] 블랙커피 [2] 설탕커피 [3] 프림커피 [4] 밀크커피"))
        if choice2 == 1:
            self.new_price = int(input("새로운 금액을 입력하세요 = "))
            self.price[0] = self.new_price
            print(self.price)
            self.menu()
            
        elif choice2 == 2:
            self.new_price = int(input("새로운 금액을 입력하세요 = "))
            self.price[1] = self.new_price
            print(self.price)
            self.menu()
            
        elif choice2 == 3:
            self.new_price = int(input("새로운 금액을 입력하세요 = "))
            self.price[2] = self.new_price
            print(self.price)
            self.menu()
            
        elif choice2 == 4:
            self.new_price = int(input("새로운 금액을 입력하세요 = "))
            self.price[3] = self.new_price
            print(self.price)
            self.menu()
            
        else:
            print("잘못된 번호입니다. 종료")
            self.menu()
            
    def wantmore(self):
        more = input("더 구매하시겠습니까?(네/ 아니오) = ")
        if more == "네":
            self.menu()
            
        elif more == "아니오":
            print("안녕히 가세요. 거스름 돈 = {}".format(self.wallet))
            self.wallet = 0
            self.menu()
        
        else:
            print("잘못된 접근입니다.")
            self.wantmore() # 물어보는 곳으로 되돌아 가기 
   

    def menu(self):
        print("="*15)
        print("커피 자동 판매기 동작")
        print("="*15)
        print("1. 블랙커피 ({}원)".format(self.price[0]))
        print("2. 설탕커피 ({}원)".format(self.price[1]))
        print("3. 프림커피 ({}원)".format(self.price[2]))
        print("4. 밀크커피 ({}원)".format(self.price[3]))
        print("="*15)
        self.money = input("투입금액 = ")
        if self.money == "manager":
            self.management()
                 # 관리 프로그램으로 들어가도록

        else:
            self.order = int(input("주문번호 = "))
            if 1 <= self.order <= 4:
                self.money = eval(self.money)
                self.wallet += self.money
                self.makeandchange()
            else:
                print("잘못된 접근입니다.")
                self.menu()
                
```
